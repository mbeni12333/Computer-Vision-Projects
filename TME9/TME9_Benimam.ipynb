````python
def split(I,reg,pred,*args):
    """ Array*Bloc*(Array*Bloc*...->bool)*... -> 4-aire
        Découpage de I en arbre 4-aire sur un critère d'écart-type minimal
    """
    x, y, width, height = reg
    
    if(pred(I, reg, *args) and width > 2 and height > 2):

        
        width_child, height_child = width//2, height//2
                

        # Nord ouest
        NO = split(I, (x, y, width_child, height_child), pred, *args)
        # Nord est
        NE = split(I, (x + width_child, y, width_child, height_child), pred, *args)
        # Sud Ouest
        SO = split(I, (x, y+height_child, width_child, height_child), pred, *args)
        # Sud Est
        SE = split(I, (x + width_child, y + height_child, width_child, height_child), pred, *args)
        
        
        return [NO, NE, SO, SE]

    return [reg]
    
def predsplit1(I,reg,*args):
    """ Array*Bloc*... -> bool
    """ 
    x, y, width, height = reg
    
#     print("x = ", x, "y = ", y)
    # variation supereir a un threshhold

    return I[y:y+height, x:x+width].std() > args[0]

def splitprof(L):
    """ 4-aire -> list[Bloc]"""
    
    if(len(L) == 1):
        return L

    childs = []
    for i in range(4):
        #print("L is = ", L[i])
        childs += splitprof(L[i])
        
    return childs

def splitdraw(LL,I):
    """ list[Bloc]*Array -> Array
        parcours de la liste dessin des régions 
    """
    I_copy = np.zeros_like(I)
    
#     cols = 5
#     rows = len(LL)//cols
    
#     fig1, f1_axes = plt.subplots(ncols=cols, nrows=rows, constrained_layout=True, figsize=(30,20))
    
    for i, reg in enumerate(LL):
        
        x, y, width, height = reg
        
        I_copy[y:y+height, x:x+width, 0] = I[y:y+height, x:x+width, 0].mean()
        I_copy[y:y+height, x:x+width, 1] = I[y:y+height, x:x+width, 1].mean()
        I_copy[y:y+height, x:x+width, 2] = I[y:y+height, x:x+width, 2].mean()
        
        if(y != 0):
        #horizontal
            I_copy[y, x:x+width, 0] = 255
            I_copy[y, x:x+width, 1] = 0
            I_copy[y, x:x+width, 2] = 0
        
        if(x != 0):
        #verticla
            I_copy[y:y+height, x, 0] = 255
            I_copy[y:y+height, x, 1] = 0
            I_copy[y:y+height, x, 2] = 0
        
        #affichage_14([RGB/255],['titre'])
#         row = i // cols 
#         col = i % cols
        
#     f1_axes[row, col].imshow(I_copy/255)
#     f1_axes[row, col].axis("off")
    plt.imshow(I_copy/255)
    plt.axis("off")
    plt.show()

def split_img_show(imgs):
    """
        dict[str*numpy]
    """
    
    for title, img in imgs.items():
        

        plt.figure(figsize=(20,10))

        #noise = random.randn(img.shape[0], img.shape[1]) * 1

        #img_noise = img + noise
        
        plt.subplot(121)
        plt.imshow(img, cmap="gray")
        plt.title(title)
        plt.axis("off")


        RGB = zeros((img.shape[0], img.shape[1],3))
        RGB[:,:,0] = img
        RGB[:,:,1] = img
        RGB[:,:,2] = img

        plt.subplot(122)
        plt.title(title + "_quadtree_split")
        split_test = split(img, (0, 0, img.shape[1], img.shape[0]), predsplit1, 20)
        split_prof_test = splitprof(split_test)
        splitdraw(split_prof_test, RGB)
````

### Question 1
Modifier la fonction ``splitprof(L)`` pour en faire une fonction ``splitprofstat(L,I)`` qui rend une liste de feuilles de l'arbre 4-aire ``L`` et telle que chaque élément de cette liste, de type ``Bloc``, est enrichi de trois scalaires, le premier étant la taille, le second la moyenne et la troisième la variance des valeurs de pixels du bloc dans l'image ``I``. Cette fonction rend alors une liste dont les éléments ont pour type formel:
```python
type Blocstat = tuple[int**4,int,float**2]
```
Les quatre premières valeurs sont celles du type ``Bloc``, la cinquième est la taille du bloc (en nombre de pixels) et les deux dernières correspondent à la moyenne et la variance calculées sur la région.


````python
# type Blocstat = tuple[int**4,int,float**2]

def splitprofstats(L,I):
    """ 4-aire*Array -> list[Blocstat] """
    
    if(len(L) == 1):
        
        x, y, width, height = L[0]
        
        taille = width*height
        moy = I[y:y+height, x:x+width].mean()
        std = I[y:y+height, x:x+width].std()**2

        return [(x, y, width, height, taille, moy, std)]

    childs = []
    for i in range(4):
        #print("L is = ", L[i])
        childs += splitprofstats(L[i], I)
        
    return childs
````

Une région, comme vu au TME, est donc une liste de blocs. Écrire le prédicat de fusion ``predmerge(b,R,*args)`` vu en TD. Cette fonction retourne ```True``` si le bloc ``b`` doit fusionner dans la région ``R``. Si telle est le cas, alors le premier élément de la liste ``R`` voit ses statistiques modifiées: elles décrivent celles de l'ensemble des blocs de la région ainsi fusionnée.

````python
def predmerge(b,R,*args):
    """ Blocstat*Region*... -> bool 
        Si fusion, R[0] est modifié
    """
    
    if(b in R):
        return False
    
    # si la variance decroit
    x, y, width, height, taille, moy, std = b
    x_reg, y_reg, width_reg, height_reg, taille_reg, moy_reg, std_reg = R[0]
    
    #if(taille > taille_reg):
    #    return False
    
    # nouvel donné de la fusion
    n = taille + taille_reg
    mu = (taille*moy + taille_reg*moy_reg)/n
    std3 = 1/n * (taille*(std+moy**2) + taille_reg*(std_reg+moy_reg**2)) - mu**2
    #print("std3 : ", (std3 - (taille*std + taille_reg*std_reg)/n))
    # evite de specefier un seuil
    critere =  (std3 - (taille*std + taille_reg*std_reg)/n) <= 0
    
    if std3 < args[0]:
        # fusioner avec la region
        R[0] = x_reg, y_reg, width_reg, height_reg, n, mu, std3
        # ajouter dans la liste des voisins
        #R.append(b)
        
        return True
    
    
    return False
    
````

### Question 3
Implémenter les fonctions ```estdans()```, ```sontvoisins()```, ```voisins()``` et finalement ```merge()``` vues en TD.

````python
def estdans(x,y,a):
    """ int*int*Blocstat -> bool 
        Teste si (x,y) est dans a 
    """
    x_a, y_a, width, height, n, m, v = a
    
    return (x >= x_a) and (x < (x_a + width)) and (y >= y_a) and (y < (y_a + height)) 


def sontvoisins(a,b):
    """ Blocstat**2 -> bool
        Teste si a et b sont voisins
    """
    
    # la region b doit étre la plus large, si ce n'est pas le cas , permute
    if (a[2]>b[2]) and (a[3]>b[3]):
        a, b = b, a
    
    # le centre de la region a
    x, y = a[0]+a[2]//2, a[1]+a[3]//2
             
    # verifier les 4 direction
    return (estdans(x+a[2], y, b) or
            estdans(x-a[2], y, b) or
            estdans(x, y+a[3], b) or
            estdans(x, y-a[3], b))

def voisins(b,K):
    """ Blocstat*list[Blocstat]->list[Blocstat]
        retourne la liste des elements de K voisins de b
    """
    return [n for n in K if(sontvoisins(b, n))]
    
    
# type Region = list[Blocstat]

def merge(S,I,pred,*args):
    """ 4-aire*Array*(Blocstat*Region*...->bool) -> list[Region]
        Fusionne les feuilles de S en une liste de régions
    """
    S = splitprofstats(S, I)
    S.sort(key=lambda x:x[4], reverse=True)
    #print(S)
    #K = [(x, y, n, m, n*m, I[x:x+n, y:y+m].mean(), I[x:x+n, y:y+m].std()) for (x, y, n, m) in S]
    K = S
    # liste des regions
    L = []
    
    # tanq qu'il reste un block dans K
    while(len(K) > 0):
        b = K[0]
        K =  K[1:]
        
        # nouvel region
        R = [b]
        
        # voisin de b
        
        N = voisins(b, K)
        N.sort(key=lambda x:x[4], reverse=True)
        # tant qu'on a pas visité tout les voisins
        while(len(N) > 0):
            
            bi = N[0]
            N = N[1:]
            
           
            if pred(bi, R, *args):
                # supprimer l'element 
                #print("merged !")
                K.remove(bi)
                
                R.append(bi)
                
                # ajouter les voisin de la regioné fusioné
                N = N + [j for j in voisins(bi, K)]
                #print("not merged !")
        # Ajouter cette nouvel region
        L.append(R)
        
    return L               
````

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def harris_corners(image):
    image_copy = image[:,:]
    #Converting the image to grayscale
    gray_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    
    #Conversion to float is a prerequisite for the algorithm
    gray_img = np.float32(gray_img)
    
    # 3 is the size of the neighborhood considered, aperture parameter = 3
    # k = 0.04 used to calculate the window score (R)
    corners_img = cv2.cornerHarris(gray_img,3,3,0.04)
    
    #Marking the corners in Green
    image_copy[corners_img>0.001*corners_img.max()] = [0,255,0]

    return image_copy
```

```python
image_house = cv2.imread("images/house2.jpg", cv2.IMREAD_COLOR)
image_rectangle = cv2.imread("images/rectangle.png")
image_carrelage = cv2.imread("images/carrelage_wikipedia.jpg")
image_jussieu = cv2.imread("images/Jussieu_wikipedia.jpg")
```

```python
harris_corners(image_jussieu)
plt.figure(figsize=(20,20))
plt.imshow(image_jussieu)
```
